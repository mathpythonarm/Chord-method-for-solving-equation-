#### Chord method for solving equation

```python
import math
import numpy as np
# enter the equation in f(x) 
def f(x):
    return np.sin(x**2)-x-1
```

```python
import matplotlib.pyplot as plt

x= np.arange(-5, 5, 0.1).reshape((-1, 1))
z= np.zeros(x.size)
plt.plot(x, f(x))
plt.plot(x,z)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
```

```python
def secant_method(f, x0, x1, tol=1e-5, n=0): # enter error in tolerance
    n += 1 
    y0, y1 = f(x0), f(x1) # calculate function values at endpoints
    
    xn = x1 - y1 * ((x1 - x0) / (y1 - y0)) # calculate next root approximation
    print(f'n={n} xn={xn} f(xn)={f(xn)}')
    
    if -tol < y1 and y1 < tol:
        
        return xn, n # check tolerance condition
    
    return secant_method(f, x1, xn, n=n) # recursive call with updated interval
```

```python
secant_method(f, -1, 0) # Enter x0 and x1 to call function and run
```
