```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
f = lambda x:  1/(9+x**3)
```

```python
a= 2 ; b= 4 ; N= 10 ; n=10
```

```python
x = np.linspace(a , b, N+1) 
```

```python
midpoints= f(x[1:-1])
trapezoid= (f( x[0] )+2*(sum(midpoints))+ f( x[N] ))*((b-a)/(2*N))
trapezoid
```

```python
 
X= np.linspace (a, b, n**2+1)
Y = f(X)
y = f(x)

plt.figure(figsize=(8, 5))
plt.plot(x , y) #plot points on each bar
plt.bar(x, y, edgecolor = 'b', alpha=0.3, width=0.01) 
plt.plot( X, Y) # plt function
```
