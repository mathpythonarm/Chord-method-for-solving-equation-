```python
import sympy as smp
from sympy import *

x = smp.symbols('x')
# input f after smp.
f= smp.cos(x)-2*x**3
f
```

```python
df= smp.diff(f, x, 1)
df
```

```python
g=x- f/df
g
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
# enter the equation in f(x) 
def f(x):
    return np.cos(x)-2*x**3

x= np.arange(-2, 2, 0.1).reshape((-1, 1))
z= np.zeros(x.size)
plt.plot(x, f(x))
plt.plot(x,z)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
```

```python
max_iter = 20  # Max iterations
x0=1 # set initial value to approximate root
root= x0

tol = 1E-5

for i in range(1, max_iter+1):
    root= g.subs(x, root).evalf()
    if abs(f.subs(x, root)) < tol:
        print(f'f(x)=0 has {root},and in this point:')
        print(f'f({root})={f.subs(x, root)}')
        print(f'iterate counter is:{i}')
        break
```
